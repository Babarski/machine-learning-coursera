5)Коррелируют ли число братьев/сестер с числом родителей/детей?
Посчитайте корреляцию Пирсона между признаками SibSp и Parch.

```python
sumSibSP = np.sum(data['SibSp'])
print sumSibSP
sumParch = np.sum(data['Parch'])
print sumParch

mSibSp = np.mean(data['SibSp'], axis=0)
print mSibSp
mParch = np.mean(data['Parch'], axis=0)
print mParch

dSibSp = []
for i in range(1, len(data['SibSp'])):
    dSibSp.append(data['SibSp'][i] - mSibSp)

print "dSibSp[0]  "+str(dSibSp[0])  

dSibSpSquare = np.square(dSibSp)
print "SibSp square "+ str(dSibSpSquare[0])

dParch = []
for i in range(1, len(data['Parch'])):
    dParch.append(data['Parch'][i] - mParch)
    
print "dParch[0] "+str(dParch[0])  

dParchSquare = np.square(dParch)
print "Parch square "+ str(dParchSquare[0])

dSibSpXdParch = []
for i in range(0, len(data['Parch'])-1):
    vir = dParch[i]*dSibSp[i]
    dSibSpXdParch.append(vir)
    
print "dSibSpXdParch[0] "+str(dSibSpXdParch[0]) 

sumdSibSpSquare = np.sum(dSibSpSquare)
print "\nsumdSibSpSquare "+ str(sumdSibSpSquare)
sumdParchSquare = np.sum(dParchSquare)
print "\nsumdParchSquare "+ str(sumdParchSquare)

sumdSibSpXdParch = np.sum(dSibSpXdParch)
print "\nsumdSibSpXdParch " + str(sumdSibSpXdParch)

sq = np.sqrt(sumdSibSpSquare*sumdParchSquare)
res = sumdSibSpXdParch/sq
print round(res,2)
```
